# Dynamic Prompts


```python
from langchain_core.prompts import PromptTemplate
# Create a template with placeholders
template = PromptTemplate.from_template(
"Summarize the topic '{topic}' in a '{tone}' tone."
)
# Now you can dynamically fill the placeholders
prompt_1 = template.format(topic="Cricket", tone="fun")
print(prompt_1)
```

# 2 Role Based Prompts

```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages([
("system", "You are an experienced {profession}."),
("human", "Tell me about {topic}."),
])
prompt_for_doctor=chat_template.format_messages(profession="Doctor",
topic="Viral Fever")
print(prompt_for_doctor)
```

# 3.Few-Shot Prompts

```python
from langchain_core.prompts import FewShotPromptTemplate, PromptTemplate

examples = [
    {"query": "I was charged twice for my subscription.", "category": "Billing Issue"},
    {"query": "The app crashes every time I try to log in.", "category": "Technical Problem"},
    {"query": "Can you explain how to upgrade my plan?", "category": "General Inquiry"},
]

example_template = PromptTemplate.from_template("Query: {query}\nCategory: {category}")

few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_template,
    prefix="Classify the following customer support tickets into one of the categories: Billing Issue, Technical Problem, General Inquiry",
    suffix="Query: {input}\nCategory:",
    input_variables=["input"]
)

final_prompt = few_shot_prompt.format(input="I forgot my password")
print(final_prompt)
```
