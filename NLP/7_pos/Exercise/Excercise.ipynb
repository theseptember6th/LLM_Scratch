   1) Extract all NOUN tokens from this story. You will have to read the file in python first to collect all the text and then extract NOUNs in a python list

   3) Print a count of all POS tags in this story





# Reading the file

```python
with open("news_story.txt","r") as f:
    text=f.read()
```

```python
text
```

```python
import spacy

spacy.prefer_gpu()
nlp = spacy.load("en_core_web_sm")
```

```python
doc=nlp(text)
noun=[]
numbers=[]
for token in doc:
    if token.pos_=="NOUN" or token.pos_=="PROPN":
        noun.append(token.text)
    elif token.pos_=="NUM":
        numbers.append(token.text)

print(noun)
print("="*10)
print(numbers)
```

```python
pos_counts=doc.count_by(spacy.attrs.POS)
print(pos_counts)
```

```python
doc.vocab[92].text
```

```python
for keys,values in pos_counts.items():
    print(doc.vocab[keys].text,values)
```
