### Pie chart
- A pie chart is a circular chart used to show percentage or proportion of each category with respect to the total.
- Each slice represents how much one item contributes to the whole.

```python
import matplotlib.pyplot as plt
```

<h3 style="color:purple">Draw pie chart to track down home expenses</h3>

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

<h3 style="color:purple">Draw pie chart as perfect circle</h3>

```python
plt.pie(exp_vals,labels=exp_labels)
plt.axis("equal") # for making perfect circle
plt.show()
```

<h3 style="color:purple">Show percentages for every pie. Specify radius to increase chart size</h3>

- autopct displays percentage on each pie
- it uses format strings
    - "%.2f%" -> 2 decimal place
    - %.1f% - > 1 decimal place
    - %d%% -> no decimal

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct="%.1f%%",shadow=True)
# shadow = True -> adding shadow effects
```

<h3 style="color:purple">Explode</h3>

- explode is used to pull out a slice
- it is a list of same length as number of slices
- value indicates how for the slice move out
- explode=[...,...,...]

```python
exp_labels
```

```python
len(exp_labels)
```

```python
plt.pie(exp_vals,labels=exp_labels,shadow=True,explode=[0,0.2,0,0,0],autopct="%.1f%%")
plt.axis("equal")
plt.show()
```


<h3 style="color:purple">counterclock and angle properties</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=45)
plt.show()
```
