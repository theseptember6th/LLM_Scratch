<h2 style="color:green" align="center">Saving chart to a file using savefig</h2>

```python
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
```

```python
plt.savefig("empty_piechart.png")
```

# it saved empty white image file because u run plt.savefig in separate code cell and plt resets everytime in new cell

# so always save in same code cell

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("Kristal_Piechart.png")
plt.show()
```

```python
from matplotlib.transforms import Bbox


plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("Kristal_Piechart_fit.png",bbox_inches="tight")
plt.show()
```

```python
from matplotlib.transforms import Bbox


plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("Kristal_Piechart_fit_inches.png",bbox_inches="tight",pad_inches=2)
plt.show()
```

```python
from matplotlib.transforms import Bbox


plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("Kristal_Piechart_fit_inches_Transparent.png",bbox_inches="tight",pad_inches=2,transparent=True)
plt.show()
```

<h2>You can also save in pdf format and to a different location on your computer by specifying absolute path<h2>

```python
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("c:/code/piechart.pdf", bbox_inches="tight", pad_inches=10, transparent=True)
```
