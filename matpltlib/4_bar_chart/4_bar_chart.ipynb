<h1 align="center", style="color:green">Matplotlib Tutorial: Bar Chart</h1>

```python

import matplotlib.pyplot as plt
import numpy as np
```

<h3>Simple bar chart showing revenues of major US tech companies</h3>

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
plt.bar(company,revenue)
```

```python
plt.bar(company,revenue, label="Revenue")

plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

<h3>Multiple Bars showing revenue and profit of major US tech companies</h3>

```python
profit=[40,2,34,12]
```

```python
plt.bar(company,revenue, width=0.4, label="Revenue")
plt.bar(company,profit, width=0.4,label="Profit")

plt.xlabel("Company")
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

####  in above bar chart,both bars are same position so to shift bars, to separate we use concept  x_pos+x_ticks

```python
x_pos=np.arange(len(company)) # change categorical to number

plt.bar(x_pos+0.2,revenue, width=0.4, label="Revenue")
plt.bar(x_pos-0.2,profit, width=0.4,label="Profit")

plt.xlabel("Company")
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')

plt.xticks(x_pos,company)# replace x_pos with categorical company variables
plt.legend()
```

<h3>Horizontal bar chart using barh function<h3>

```python
plt.barh(x_pos,revenue, label="Revenue")

plt.yticks(x_pos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```
