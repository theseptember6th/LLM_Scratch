

# **Naive Bayes Classifier – Part 1**

### **1. Introduction**

* Naive Bayes is a **Machine Learning algorithm** used for **classification** tasks.
* Based on **Bayes’ Theorem** from probability theory.
* It is called **“Naive”** because it assumes that all features are **independent** of each other.
* Despite being a simple assumption, it works surprisingly **well and efficiently**.

---

### **2. Basic Probability Concepts**

* Example 1: Tossing a coin →
  Probability of Head or Tail = **1/2 = 0.5 (50%)**

* Example 2: Picking a Queen from 52 cards →
  4 Queens → Probability = **4/52 = 1/13**

* Example 3: Conditional Probability
  If you already know the card is a **Diamond**,
  then Probability of it being a **Queen** = **1/13**

  Represented as:
  [
  P(A|B) = \text{Probability of A given B has already occurred}
  ]

---

### **3. Bayes’ Theorem**

Bayes theorem helps to find the probability of an event based on prior knowledge of other related events.

[
P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
]

Example: In the card problem,

* (A): Card is Queen
* (B): Card is Diamond
  You can find (P(A|B)) using the above formula.

---

### **4. Applying Bayes’ Theorem in Real Life (Titanic Example)**

* Dataset: Titanic passengers with features like
  **Sex, Class, Age, Cabin, Fare, etc.**
* Goal: Predict whether a person **survived or not**.
* We can use Bayes Theorem to find:
  [
  P(\text{Survived} | \text{Features})
  ]
* **Naive assumption:** All features (sex, age, class, etc.) are **independent**.
  → That’s why it’s called **Naive Bayes**.

### **5. Where Naive Bayes is Used**
