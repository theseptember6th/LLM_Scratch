```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    iterations=10000
    n=len(x)
    learning_rate=0.08 # if you put learning rate =0.1 the cost function overshoots and increases instead of decreasing
    

    for i in range(iterations):
        y_predicted=m_curr * x + b_curr
        cost= (1/n) * sum([ val**2 for val in (y -y_predicted)]) #Sum of squared error SSE

        # cost=np.mean((y-y_predicted)**2) # is more efficient
        md= -(2/n) * sum(x * (y - y_predicted)) # partial derivative of cost function wrt to slope
        bd=-(2/n)* sum(y - y_predicted) # partial derivative of cost function wrt to intercept
        m_curr=m_curr - learning_rate * md
        b_curr=b_curr - learning_rate * bd
        print(f"m {m_curr} b {b_curr} iteration {i} costfunction {cost} y_predicted {y_predicted}")

    return m_curr,b_curr
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
m_final,b_final=gradient_descent(x,y)
```

```python
# Plot data points
plt.scatter(x, y, color='red', marker='+', linewidth=5)
plt.plot(x,m_final*x+b_final,color="blue")
plt.show()
```
