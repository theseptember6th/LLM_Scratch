

# **Machine Learning - Hyperparameter Tuning (GridSearchCV)**

### **1. Choosing a Model**

* When solving a machine learning problem (e.g., classifying the Iris flower dataset), the first step is selecting a suitable model.
* Example: Using SVM to classify iris flowers based on petal and sepal measurements.

---

### **2. Hyperparameters**

* Hyperparameters are settings of the model that we must choose **before training**.
* For SVM, common hyperparameters include:

  * **Kernel type**: `linear`, `RBF`, etc.
  * **C**: Regularization parameter.
  * **Gamma**: Kernel coefficient.
* Choosing the best hyperparameters is called **hyperparameter tuning**.

---

### **3. Problem with Random Parameters**

* Training an SVM with randomly chosen parameters may give variable scores depending on the train/test split.
* Example: Score may be 95% one time, 91% the next.

---

### **4. K-Fold Cross-Validation**

* Solution: Use **k-fold cross-validation** to get stable performance estimates.
* Process:

  1. Split data into `k` folds (e.g., 5).
  2. Use one fold as test set and remaining as training set.
  3. Train the model and calculate the score.
  4. Repeat for all folds.
  5. Take the average score.
* This ensures all samples are tested and reduces score variability.

---

### **5. Manual Hyperparameter Search**

* One way is to try combinations of hyperparameters using nested loops.
* Example for SVM:

  ```python
  kernels = ['linear', 'rbf']
  C_values = [1, 10, 20]
  for k in kernels:
      for c in C_values:
          model = SVC(kernel=k, C=c)
          scores = cross_val_score(model, X, y, cv=5)
          print(k, c, scores.mean())
  ```
* Limitations:

  * Manual and repetitive.
  * Hard to scale if multiple hyperparameters exist.

---

### **6. GridSearchCV**

* `GridSearchCV` automates hyperparameter tuning with cross-validation.
* Steps:

  1. Import:

     ```python
     from sklearn.model_selection import GridSearchCV
     ```
  2. Define classifier and parameter grid:

     ```python
     param_grid = {'C':[1,10,20], 'kernel':['linear','rbf']}
     clf = GridSearchCV(SVC(gamma='auto'), param_grid=param_grid, cv=5)
     clf.fit(X, y)
     ```
  3. Check results:

     * `clf.best_params_` → best hyperparameter combination.
     * `clf.best_score_` → best cross-validation score.
* `GridSearchCV` tries **all combinations** of the parameters efficiently.

---

### **7. RandomizedSearchCV**

* Problem with GridSearchCV: Computationally expensive for large datasets or many parameters.
* Solution: `RandomizedSearchCV` tries a **random subset of parameter combinations**.
* Example:

  ```python
  from sklearn.model_selection import RandomizedSearchCV
  clf = RandomizedSearchCV(SVC(), param_distributions=param_grid, n_iter=2, cv=5)
  clf.fit(X, y)
  ```
* Useful when computation resources are limited.

---

### **8. Choosing the Best Model**

* You can test multiple classifiers (e.g., SVM, Random Forest, Logistic Regression) using GridSearchCV.
* Steps:

  1. Define a dictionary of classifiers and their parameter grids:

     ```python
     models = {
         'SVM': {'model': SVC(), 'params': {'C':[1,10], 'kernel':['linear','rbf']}},
         'RandomForest': {'model': RandomForestClassifier(), 'params': {'n_estimators':[5,10,15]}},
         'LogisticRegression': {'model': LogisticRegression(), 'params': {'C':[1,10]}}
     }
     ```
  2. Loop through each model and use GridSearchCV to find best parameters.
  3. Compare the scores to select the best model for your problem.
* Example conclusion: For the Iris dataset, **SVM achieved 98% score**, better than Random Forest and Logistic Regression.

---

### **9. Summary**

* Hyperparameter tuning is crucial to maximize model performance.
* **Manual tuning** is possible but not efficient.
* **GridSearchCV**: Exhaustive search over parameter grid with cross-validation.
* **RandomizedSearchCV**: Random search for faster tuning when computation is limited.
* **Best model selection**: Test multiple models with hyperparameter tuning to select the one with the highest score.
