## To predict the value at 2020

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import linear_model
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
```

```python
df.head()
```

```python
reg=linear_model.LinearRegression()
```

```python
X=df[["year"]] # independent variable have to be 2D array/list
y=df["per capita income (US$)"] # dependent variable have to be 1D array/list
```

```python
reg.fit(X,y)
```

```python
slope=reg.coef_
print(slope)
```

```python
intercept1=reg.intercept_
print(intercept1)
```

```python
plt.scatter(X,y,color='red',marker="*")
```

```python
plt.xlabel("year")
plt.ylabel("per capita income")
plt.scatter(X,y,color='red',marker="*")
plt.plot(X,reg.predict(X))
```

```python
to_predict_value=2020 # to predict
reg.predict([[to_predict_value]])
```
