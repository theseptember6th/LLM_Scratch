# **Machine Learning Tutorial – Logistic Regression (Binary Classification)**

## **1. Introduction**

* Logistic Regression is used to solve **classification problems**.
* **Linear regression** predicts **continuous values** (e.g., home prices, stock prices, weather).
* Classification predicts **categorical values** (e.g., spam email yes/no, insurance purchase yes/no, voting preference).

---

## **2. Types of Classification**

1. **Binary Classification**: Outcome has **two possible categories**.

   * Example: Will the customer buy insurance? (Yes/No)
2. **Multiclass Classification**: Outcome has **more than two categories**.

   * Example: Predicting which party a person will vote for.

---

## **3. Example Problem**

* Task: Predict if a customer will buy an insurance product based on **age**.
* Observed Pattern:

  * Younger people → less likely to buy.
  * Older people → more likely to buy.
* Data is visualized using a **scatter plot**.

---

## **4. Why Not Linear Regression for Classification**

* Linear regression can be applied, but problems arise:

  * Example: Using a cutoff like 0.5 to predict yes/no.
  * Outliers (very old or young customers) may cause incorrect predictions.
* Linear regression line may not separate categories well.

---

## **5. Sigmoid Function**

* Converts linear output into a **range between 0 and 1**.
* **Equation**:
  [
  \sigma(z) = \frac{1}{1 + e^{-z}}
  ]

  * (e) is Euler’s number (~2.718)
* Produces an **S-shaped curve** (logistic curve).
* Logistic regression formula:

  * Feed linear equation (z = mx + b) into sigmoid function to get probabilities.
* Outcome:

  * Close to 0 → category 0 (e.g., No)
  * Close to 1 → category 1 (e.g., Yes)

---

## **6. Implementation in Python (Using Scikit-Learn)**

### **6.1 Import Libraries**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

### **6.2 Load Data**

* CSV file contains **age** and **insurance purchase (0 or 1)**.

```python
df = pd.read_csv("insurance.csv")
```

### **6.3 Data Visualization**

* Scatter plot to see distribution:

  * X-axis → Age
  * Y-axis → Insurance purchase (0/1)
* Observation: Probability of buying insurance increases with age.

### **6.4 Train-Test Split**

```python
X = df[['age']]  # multi-dimensional array
y = df['insurance']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9)
```

### **6.5 Train Logistic Regression Model**

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

### **6.6 Make Predictions**

```python
predictions = model.predict(X_test)
```

### **6.7 Model Accuracy**

```python
accuracy = model.score(X_test, y_test)
```

* For small datasets, score may appear **perfect (1.0)**.
* For larger datasets, expect **some errors**.

### **6.8 Predict Probabilities**

```python
probabilities = model.predict_proba(X_test)
```

* Shows probability of each class (0 or 1).
* Example: Age 18 → low probability to buy insurance.
* Example: Age 40 → high probability to buy insurance.
