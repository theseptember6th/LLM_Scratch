# Machine Learning Tutorial: Dummy Variables & One Hot Encoding Explained

### Context:

You have a dataset with home prices. The features (inputs) include:

* **Area of the home (in square feet)**
* **Township name** (e.g., Mondro, Westincor, Robinsville)

Your goal is to **predict the price** of a house given these inputs.

---

## Problem With Categorical Data (Township Name)

* Machine learning models **work well with numbers** but **not directly with words or categories** like town names.
* Township is a **categorical variable** because it holds categories (town names), not numbers.

### What to do?

#### 1. Label Encoding (Integer Encoding)

* Convert each town name into an integer, like:

  * Mondro → 0
  * Westincor → 1
  * Robinsville → 2
* This looks easy but it causes a problem:
  The model **will treat these numbers as if they have order or meaning**, for example:

  * "0 < 1 < 2" or "0 + 1 = 2"
    But town names don’t have any order. They are just different categories.

This can confuse the model and lead to wrong assumptions.

---

## Types of Categorical Variables

* **Nominal variables**: Categories **without any order or ranking**
  Examples: Town names, colors, gender (male/female)
  → These should NOT be encoded as integers that imply order.

* **Ordinal variables**: Categories **with a meaningful order**
  Examples: Education level (Bachelor < Master < PhD), customer satisfaction (Dissatisfied < Neutral < Satisfied)

In your case, **township is a nominal variable**.

---

## Solution: One Hot Encoding

* One hot encoding **creates a new binary (0/1) column for each category**.
* For each sample (row), only one of these new columns is 1, rest are 0.

Example for 3 towns:

| Township    | Mondro | Westincor | Robinsville |
| ----------- | ------ | --------- | ----------- |
| Mondro      | 1      | 0         | 0           |
| Westincor   | 0      | 1         | 0           |
| Robinsville | 0      | 0         | 1           |

Now, the model sees **only numbers** but no implied order between towns.

---

## Dummy Variables and Dummy Variable Trap

* These new one-hot encoded columns are called **dummy variables**.
* When you have N categories, one hot encoding creates N columns.
* However, one of these columns is **redundant** because it can be perfectly predicted from the others (if all are 0, it must be the last one).
* This redundancy leads to something called **dummy variable trap**, which causes **multicollinearity** — a statistical problem that messes up your model.

**Solution:**

* Drop one dummy variable column (e.g., drop 'Mondro').
* The model can still figure out the missing category when all other dummy variables are 0.

---

## How to Do This in Python with Pandas

1. Use `pandas.get_dummies()` on the "town" column:

   ```python
   dummies = pd.get_dummies(df['town'])
   ```
2. Concatenate these dummy columns with your original data:

   ```python
   merged = pd.concat([df, dummies], axis=1)
   ```
3. Drop the original "town" column (because it's text and useless for the model now):

   ```python
   final = merged.drop(['town', 'Mondro'], axis=1)  # Dropping 'Mondro' dummy to avoid dummy variable trap
   ```
4. Now you have only numeric columns in `final`, perfect for linear regression.

---

## Train a Linear Regression Model

* **X (input features)** = all columns except "price"
* **Y (target)** = "price"
* Train your model:

  ```python
  model = LinearRegression()
  model.fit(X, Y)
  ```
* Make predictions with new data:

  * Provide area and dummy variables for town (1 or 0).
  * Example: For a house in Robinsville, `Mondro=0`, `Robinsville=1`.

---

## Using Scikit-learn’s LabelEncoder + OneHotEncoder

* Another way is to use `LabelEncoder` to convert town names into integers.
* Then use `OneHotEncoder` on the label encoded data to get dummy variables.
* Important: Specify which column is categorical in `OneHotEncoder`.
* Also drop one dummy variable column to avoid dummy variable trap.

This way, you automate the process and can handle more complex pipelines.

---

## Model Evaluation

* Use `.score(X, Y)` method on your model to get accuracy (R² score for regression).
* Score near 1 means model predicts very well.
* Lower score means the model is not fitting well.

## Summary

| Step                     | Why?                                                               | How?                                        |
| ------------------------ | ------------------------------------------------------------------ | ------------------------------------------- |
| Categorical data problem | ML models only understand numbers                                  | Convert categories to numbers               |
| Label Encoding           | Converts text to integers but implies order                        | Use with ordinal data only                  |
| One Hot Encoding         | Represents categories as separate binary columns, no order implied | `pd.get_dummies()` or sklearn OneHotEncoder |
| Dummy Variable Trap      | Multicollinearity due to redundant dummy vars                      | Drop one dummy variable column              |
| Train Linear Regression  | Predict price based on area and town                               | Use sklearn LinearRegression                |
| Model accuracy           | Check model performance                                            | Use `.score()` method                       |
