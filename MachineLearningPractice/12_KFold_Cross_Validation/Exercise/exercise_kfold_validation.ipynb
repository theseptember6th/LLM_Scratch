Use iris flower dataset from sklearn library and use cross_val_score against following
models to measure the performance of each. In the end figure out the model with best performance,
1. Logistic Regression
2. SVM
3. Decision Tree
4. Random Forest

```python
from sklearn.datasets import load_iris

iris=load_iris()
```

```python
dir(iris)
```

# doing manually

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

folds=StratifiedKFold(n_splits=3)

def get_score(model,X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    return model.score(X_test,y_test)

scores_l=[]
scores_svm=[]
scores_d=[]
scores_rf=[]


for train_index,test_index in folds.split(iris.data,iris.target):
    X_train,y_train=iris.data[train_index],iris.target[train_index]
    X_test,y_test=iris.data[test_index],iris.target[test_index]

    scores_l.append(get_score(LogisticRegression(),X_train,X_test,y_train,y_test))
    scores_d.append(get_score(DecisionTreeClassifier(),X_train,X_test,y_train,y_test))
    scores_svm.append(get_score(SVC(),X_train,X_test,y_train,y_test))
    scores_rf.append(get_score(RandomForestClassifier(),X_train,X_test,y_train,y_test))

```

# directly doing using sklearn's cross_val_score

```python
from sklearn.model_selection import cross_val_score
```

**Logistic Regression**

```python
scores=cross_val_score(LogisticRegression(),iris.data,iris.target,cv=3)
scores.mean()
```

**Decision Tree**

```python
scores=cross_val_score(DecisionTreeClassifier(),iris.data,iris.target,cv=3)
scores.mean()
```

**Support Vector Machine (SVM)**

```python
scores=cross_val_score(SVC(),iris.data,iris.target,cv=3)
scores.mean()
```

**Random Forest**

```python
scores=cross_val_score(RandomForestClassifier(),iris.data,iris.target,cv=3)
scores.mean()
```

**Best score so far is from SVM: 0.96**
