## **Machine Learning Tutorial: Training and Testing Data**

**1. Why Split Data?**

* Using the entire dataset for training is **not a good strategy**.
* Split the dataset into:

  * **Training set** → to train the model
  * **Test set** → to test the model’s performance on **unseen data**
* Example: Train on 8 samples, test on 2 → gives idea of model accuracy.

**2. Dataset Used**

* **BMW car prices dataset**
* Features:

  * **Independent variables (X):** Mileage, Age
  * **Dependent variable (Y):** Selling Price
* Visualize relationships using **scatter plots**:

  * Mileage vs Selling Price → roughly linear
  * Age vs Selling Price → roughly linear
* Linear regression is suitable based on visualization.

**3. Preparing Data**

* X = [Mileage, Age]
* Y = [Selling Price]

**4. Train-Test Split (scikit-learn)**

* **Import:** `from sklearn.model_selection import train_test_split`
* **Usage:**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

* **Parameters:**

  * `test_size` → fraction of data for testing (e.g., 0.2 = 20%)
  * `train_size` → fraction for training (optional)
  * `random_state` → ensures same random split every time

* Output: 4 datasets → `X_train`, `X_test`, `Y_train`, `Y_test`

* Random samples are chosen, not first 80% of data.

**5. Training the Model**

* **Import Linear Regression:**

```python
from sklearn.linear_model import LinearRegression
```

* Create model object: `regressor = LinearRegression()`
* Train model: `regressor.fit(X_train, Y_train)`

**6. Making Predictions**

* Predict on test data:

```python
Y_pred = regressor.predict(X_test)
```

* Compare `Y_pred` with `Y_test`

**7. Checking Accuracy**

* Use `score()` method:

```python
accuracy = regressor.score(X_test, Y_test)
```

* Example result: 89% accuracy

**8. Notes:**

* Change train-test ratio by modifying `test_size` (e.g., 0.3 → 70% train, 30% test).
* Use `random_state` to get consistent train/test split across multiple runs.
* Visualization helps to decide which model to use.

The choice of `random_state=10` (or any specific number) is **arbitrary**. Its purpose is **not about the number itself**, but about **making the split reproducible**.

Here’s why:

1. **Without `random_state`**

   * Each time you call `train_test_split`, it randomly selects different samples for training and testing.
   * This means your `X_train` and `X_test` change every run, and your results (like accuracy) might slightly change each time.

2. **With `random_state`**

   * Setting `random_state` to a fixed integer (e.g., 10, 42, 0) **freezes the randomness**.
   * Every time you run the code with that value, you get the **same train-test split**, making results consistent and easier to debug or compare.

3. **Why 10?**

   * The number itself doesn’t matter. You could use 0, 1, 42, 123, etc.
   * People often pick **any number they like**. Some choose `42` as a “fun convention” because of *The Hitchhiker’s Guide to the Galaxy*.

✅ **Summary:** `random_state` ensures reproducibility. The value 10 is just an example; any fixed integer works the same way.
