In exercise folder (same level as this notebook on github) there is **hiring.csv**. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates,


**2 yr experience, 9 test score, 6 interview score**

**12 yr experience, 10 test score, 10 interview score**


<h3>Answer<h3>

53713.86 and 93747.79

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
from sklearn import linear_model
```

```python
df=pd.read_csv('hiring.csv')
```

```python
df
```

```python
print(df.columns)
```

```python
X=df[["experience","test_score(out of 10)","interview_score(out of 10)"]]
```

```python
y=df["salary($)"]
```

## Now its time for data cleaning


```python
df["test_score(out of 10)"].median()
```

```python
mean_value=df["test_score(out of 10)"].mean()
print(mean_value)
```

```python

import math
mean_value=math.floor(mean_value)
print(mean_value)
```

```python
df["test_score(out of 10)"].mode()[0]
```

```python
df["test_score(out of 10)"]=df["test_score(out of 10)"].fillna(mean_value)
```

```python
df
```

## for experience its little tricky,we need module word2number

```python
from word2number import w2n
print(w2n.word_to_num('two point three'))
```

```python
df2=df.dropna()
```

```python
df2
```

```python
for value in df2["experience"]:
    value=w2n.word_to_num(value)
```

```python
df2
```

## value ac†s as temporary variable so it wont work

```python
df2["experience"]=df2["experience"].apply(w2n.word_to_num)
```

```python
df2
```

```python
X=df2[["experience","test_score(out of 10)","interview_score(out of 10)"]]
y=df2["salary($)"]
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(X,y)
```

**2 yr experience, 9 test score, 6 interview score**

**12 yr experience, 10 test score, 10 interview score**

```python
x1=[[2,9,6]]
x2=[[12,10,10]]
```

```python
predictedValue1=reg.predict(x1)
print(predictedValue1)
predictedValue2=reg.predict(x2)
print(predictedValue2)
```
