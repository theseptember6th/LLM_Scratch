## 📘 Machine Learning – Decision Tree

### 🔹 Why use Decision Tree?

* Some datasets can’t be separated by a **single straight line** (like in Logistic Regression).
* Example: when points of two classes are scattered in multiple regions.
* Decision Tree works by **repeatedly splitting** the dataset into smaller pure groups (called nodes).
* Hence, we use Decision Tree when:

  * Data is **non-linear** or **complex**.
  * We want a **model that’s easy to interpret** (like “if-else” rules).

---

## 🌳 Concept of Decision Tree

A **Decision Tree** divides the data into smaller subsets by asking questions on features.

Example:

```
Is company = Facebook? → Yes → Salary > 100k
                     → No → check job title
```

Each split tries to **reduce impurity** (make groups purer).

---

## 🧮 Mathematics Behind It

### 1️⃣ **Entropy (H)**

* **Definition:** A measure of **randomness** or **disorder** in the data.
* Used to decide **how pure a node** is.

Formula:

[
H(S) = - \sum_{i=1}^{n} p_i \log_2(p_i)
]

where

* ( p_i ) = proportion of class *i* in the subset
* Higher entropy → more mixed (impure)
* Lower entropy → more pure

👉 Example:

| Class | Count | p    | Calculation |
| ----- | ----- | ---- | ----------- |
| Yes   | 6     | 6/10 |             |
| No    | 4     | 4/10 |             |

[
H = -\left(\frac{6}{10}\log_2\frac{6}{10} + \frac{4}{10}\log_2\frac{4}{10}\right)
= 0.97
]

So, entropy ≈ **0.97 (high impurity)**

If all are same class → ( H = 0 ) (pure node)

---

### 2️⃣ **Information Gain (IG)**

* Measures how much **entropy decreases** after splitting the data.
* We want **high information gain** → better split.

Formula:

[
IG(S, A) = H(S) - \sum_{v \in Values(A)} \frac{|S_v|}{|S|} H(S_v)
]

where

* ( S ): original dataset
* ( A ): attribute (feature)
* ( S_v ): subset after splitting on value ( v )

👉 Example (simplified):

If splitting by *Company* gives

* Entropy before split = 0.97
* Weighted avg entropy after split = 0.2

Then
[
IG = 0.97 - 0.2 = 0.77
]

→ Company gives high information gain → **best feature to split first**.

---

### 3️⃣ **Gini Impurity (G)**

Used as an alternative to entropy (default in `DecisionTreeClassifier`).

Formula:

[
G = 1 - \sum_{i=1}^{n} p_i^2
]

👉 Example:
For 6 yes, 4 no:

[
G = 1 - [(0.6)^2 + (0.4)^2] = 0.48
]

If pure (all same class), ( G = 0 ).

---

### 🔹 Choosing the Split

At each node:

1. Compute **Entropy or Gini** for each feature.
2. Compute **Information Gain** (or reduction in Gini).
3. Select the feature with **highest Information Gain** (or lowest Gini).
4. Split data based on that feature.
5. Repeat until:

   * Nodes are pure, or
   * No further split improves purity.

---

## 🧩 Example Problem

Predict if **Salary > $100k** using features:
`Company`, `Job`, `Degree`.

Steps:

1. Calculate entropy of whole dataset.
2. Try splitting by “Company”.
3. For each company group, calculate new entropy.
4. Find Information Gain.
5. Repeat for “Job” and “Degree”.
6. The feature giving highest IG → used for first split.

👉 In example:

* Splitting by **Company** gave lowest entropy → best first split.
* Next splits used **Job**, then **Degree**.

---

## ⚙️ Implementation Steps (in Python)

```python
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import pandas as pd

# Step 1: Load dataset
df = pd.read_csv("salaries.csv")

# Step 2: Encode categorical columns
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()

df['company_n'] = le_company.fit_transform(df['company'])
df['job_n'] = le_job.fit_transform(df['job'])
df['degree_n'] = le_degree.fit_transform(df['degree'])

# Step 3: Prepare input and target
inputs = df[['company_n', 'job_n', 'degree_n']]
target = df['salary_more_than_100k']

# Step 4: Create and train model
model = tree.DecisionTreeClassifier(criterion='entropy')
model.fit(inputs, target)

# Step 5: Predict
model.predict([[2, 1, 2]])  # Example input
```

---

## ✅ When to Use Decision Tree

Use it when:

* Data is **nonlinear** or **mixed (categorical + numeric)**.
* You want **if-else style rules** or **interpretability**.
* You can tolerate some overfitting (or plan to prune).

Avoid if:

* Data is very noisy or small → may overfit easily.
