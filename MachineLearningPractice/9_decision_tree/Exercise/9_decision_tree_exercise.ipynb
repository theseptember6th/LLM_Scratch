```python
df.drop(["PassengerId","Name","Parch","Ticket","Cabin","Embarked"],axis="columns",inplace=True)
```

```python
df
```

```python
inputs=df.drop(["Survived"],axis="columns")
inputs
```

```python
inputs.Sex=inputs.Sex.map({"male":1,"female":2})
```

```python
inputs
```

# but the good way is to use LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
inputs["Sex"]=le.fit_transform(inputs["Sex"])
inputs
```

```python
inputs.info()
```

# so age has some null values

```python
age_mean=inputs["Age"].mean()
```

```python
inputs["Age"]=inputs["Age"].fillna(inputs["Age"].mean())
```

```python
inputs
```

```python
inputs.info()
```

```python
target=df["Survived"]
```

```python
target
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2,random_state=10)
```

```python
from sklearn import tree

model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```
