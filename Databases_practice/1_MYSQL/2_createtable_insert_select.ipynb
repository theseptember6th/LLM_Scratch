````python
create_table_query="""
CREATE TABLE person(
    name VARCHAR(50),
    age SMALLINT UNSIGNED,
    person_id INT AUTO_INCREMENT,
    PRIMARY KEY(person_id)
)
"""
my_cursor.execute(create_table_query)
````

### 4. Viewing Table Structure (DESCRIBE)

To see the columns and their data types, use the `DESCRIBE` command. 
- `DESCRIBE table_name`

This confirms the structure we created.

````python
my_cursor.execute("DESCRIBE person")
# Loop through the cursor to print results
for column in my_cursor:
    print(column)
````

### 5. Inserting Data into the Table

We insert data by specifying the columns we are filling (excluding the auto-increment `person_id`).`person_id` will be automatically generated.

### Important: String Formatting (Safe Practice)
Do **not** hardcode values directly into the SQL string.
Example :
### Code
```python
# BAD PRACTICE
insert_query = "INSERT INTO person (name, age) VALUES ("kristal",21)"
```

Instead, use `%s` as placeholders and pass a tuple of values. This is safer and allows variable usage.

````python
# Recommended
#SQL with placeholders
insert_query="INSERT INTO person (name,age) VALUES (%s,%s)"

#Data to insert (as a tuple)
data=("kristal",21)

#Execute
my_cursor.execute(insert_query,data)


#Important:Commit the changes to save them permanently
db.commit()
````

````python
# adding another row
data=("Biraj",19)
my_cursor.execute(insert_query,data)
db.commit()
````

## 6. Selecting (Reading) Data

To retrieve all data from the table, use the `SELECT` statement.

````python
#select all columns from the person table
select_query="SELECT * FROM person"
#execute the query using cursor
my_cursor.execute(select_query)
# fetch and print all rows
for row in my_cursor:
    print(row)
````
