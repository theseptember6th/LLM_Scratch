```python
# to check if mysql is working or not
import mysql
```

### Step 1: Import the Connector and Connect



```python
# 1. Import the MySQL Connector library
import mysql.connector

# 2. Establish a connection to the MySQL server
# We are connecting to the server itself, not a specific database yet.
db = mysql.connector.connect(
    host="localhost",   # The server is running on our own machine
    user="root",        # The username we set up (e.g., 'root' or 'Tim')
    password="gpt67rSy"     # The password for the user
)
# If this runs without error, the connection was successful.
print("Connected to MySQL Server successfully!")
```

### Step 2: Create a Cursor and Execute a Query



#### A cursor is like your hand that moves through the database, allowing you to send commands and fetch results.



```python
# 1.Create a cursor object
my_cursor=db.cursor()

#2.Write an SQL query to create a new database
#The SQL command:CREATE DATABASE database_name;

database_name="Code_with_tim_test_database"
sql_query=f"CREATE DATABASE {database_name}"

#3.Execute the query using the cursor
my_cursor.execute(sql_query)

print(f"Database {database_name} created sucessfully ")
```

### (optional)Step 3: Close the initial connection---

### We need to reconnect to the specific database.

```python
db.close()
print("Closed connection to server.")
```

### STEP 3.1:Connect to the New Database

##### Now that the database exists, we need to modify our connection to use it. This makes our subsequent queries target this specific database.

```python
import mysql.connector

#Modify the connection to include the database name
database_name="Code_with_tim_test_database"
db=mysql.connector.connect(
    host="localhost",
    user="root",
    password="gpt67rSy",
    database=database_name, # <-- We are now connecting to the new database
    buffered=True
)
if db.is_connected():
    print(f"Successfully connected to the '{db.database}' database.")

#We can now create another cursor to work within this database
my_cursor=db.cursor()

# You can now execute queries that affect this database,
# like creating tables, inserting data, etc.
```

### Step4- clean up:close the cursor and connection

```python
my_cursor.close()
db.close()
print("All connections closed")
```
