```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
search_term=input("What product do you want to search for? ")
```

```python
url=f"https://www.newegg.com/p/pl?d={search_term}&N=4131"
```

```python
page=requests.get(url).text
```

```python
doc=BeautifulSoup(page,"html.parser")
```

```python
page_text=doc.find(class_="list-tool-pagination").strong
page_text
```

```python
page_text.text
```

```python
pages=str(page_text).split("/")
pages
```

```python
pages=str(page_text).split("/")[-2]
pages
```

```python
pages=int(str(page_text).split("/")[-2].split(">")[-1][:-1])
pages
```

```python
items_found={}
```

```python
for page in range(1,pages+1):
    url=f"https://www.newegg.com/p/pl?N=4131&d=4080&page={page}"
    page=requests.get(url).text
    doc=BeautifulSoup(page,"html.parser")
    div=doc.find(class_="item-cells-wrap border-cells short-video-box items-list-view is-list")
    items=div.find_all(text=re.compile(search_term))
    for item in items:
        parent=item.parent
        if parent.name !="a":
            continue
        link=parent["href"]

        next_parent=item.find_parent(class_="item-container position-relative")
        price=next_parent.find(class_="price-current").strong.string
        items_found[item]={"price":int(price.replace(",","")),"link":link}
```

```python
print(items_found)
```

```python
sorted_items=sorted(items_found.items(), key=lambda x:x[1]["price"])
```

```python
for item in sorted_items:
    print(item[0])
    print(f'${item[1]["price"]}')
    print(item[1]["link"])
```
