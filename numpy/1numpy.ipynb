# Time

```python
size=1000
```

```python
l1=list(range(size))
l2=list(range(size))
start=time.time()
result=[x+y for x,y in zip(l1,l2)]
print((time.time()-start)*1000)
```

```python
a1=np.arange(size)
a2=np.arange(size)
start=time.time()
result=a1+a2
print((time.time()-start)*1000)
```
