- provides simple functions to save array data into text files and read data back from text files
- This is useful when we want to store array data permanently or share it with other programs.

```python
import numpy as np
```

```python
array_x=np.array([[1,2,3],[4,5,6]],dtype=np.int32)
array_x
```

### saving numpy array to a text file

```python
np.savetxt("array_x_saved_file.txt",array_x)
```

### loading numpy array from a txt file 

```python
array_loaded=np.loadtxt("array_x_saved_file.txt")
array_loaded
```
