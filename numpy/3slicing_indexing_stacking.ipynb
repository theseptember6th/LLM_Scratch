##### Vertical stacking of arrays (vstack)
- vstack() stacks arrays vertically(one above another)

```python
a=np.arange(6).reshape(3,2) #- arange + reshape for faster numpy array declaration
a
```

```python
b=np.arange(6,12).reshape(3,2)
b
```

```python
np.vstack((a,b))#vstack() stacks arrays vertically(one above another)
```

#### Horizontal stacking of arrays(hstack)
- hstack() stacks arrays side by side

```python
np.hstack((a,b))
```

### Horizontal splitting of arrays(hsplit)
- hsplit() divides an array into equal vertical parts(side by side splitting)

```python
a=np.arange(30).reshape(2,15)
a
```

```python
np.hsplit(a,3) # 3 partitions
```

### vertical splliting of arrays (vsplit)
- vsplit() divied arrays row-wise (one above another splitting)

```python
np.vsplit(a,2) # 2 partitions
```

### Boolean indexing in numpy 
- Boolean indexing allows selecting element based on conditions

```python
a=np.arange(12).reshape(3,4)
a
```

```python
a>4
```

```python
b=a>4
b
```

### Extracting elements using boolean index
- when boolean array is used as an index,Numpy only returns the 1D of True Elements


```python
a[b]
```

```python
a[a>4]
```

#### Replacing values using Boolean indexing
- we can modify array values using conditions

```python
a[a>4]=-1 # those values greater than 4 gets replaced
a
```
