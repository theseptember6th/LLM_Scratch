```python
from transformers import pipeline  # For working with pre-trained models
from langchain_huggingface import HuggingFacePipeline  # LangChain integration
from langchain_core.prompts import PromptTemplate  # For creating prompts
import torch  # Deep learning framework
```

```python
print(torch.cuda.is_available())
```

```python
print(torch.cuda.get_device_name(0))
```

```python


model = pipeline(
    "text-generation",
    model="mistralai/Mistral-7B-v0.1",
    device_map="auto",
    torch_dtype=torch.float16,
    load_in_8bit=True,
    max_new_tokens=256,
)
```

```python
llm=HuggingFacePipeline(pipeline=model)
```

```python
template=PromptTemplate.from_template(
    "Explain {topic} in detail for a {age} year old to understand."
)
```

```python
chain=template | llm
topic=input("Topic: ")
age=input("Age: ")
```

```python
response=chain.invoke({
    "topic":topic,
    "age":age
})
print(response)
```
