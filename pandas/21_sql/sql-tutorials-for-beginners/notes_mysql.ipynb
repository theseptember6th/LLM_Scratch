# SQL Tutorial Notes (Beginner Friendly)

---

## 1. Introduction to SQL

### Explanation

SQL (Structured Query Language) is used to **store, retrieve, manipulate, and analyze data** in databases.
It is a **must-have skill** for:

* Data Analysts
* Data Scientists
* Software Engineers
* Backend Developers

Most real-world applications (IMDb, Amazon, banking apps) use SQL in the backend.

---

## 2. Installing MySQL (Windows Overview)

### Explanation

MySQL is a popular **relational database management system (RDBMS)**.
MySQL Workbench is a **GUI tool** used to write and run SQL queries.

**Steps (summary):**

1. Uninstall old MySQL versions (if any)
2. Download MySQL Installer from MySQL Community Downloads
3. Choose **Custom installation**
4. Install:

   * MySQL Server
   * MySQL Workbench
5. Set root password (example: `root`)
6. Finish installation and open MySQL Workbench

## 6. SELECT Query (Basic)

### Explanation

`SELECT` is used to **retrieve data** from a table.

### Code

```sql
SELECT * FROM movies;
```

### Output

Returns **all rows and all columns** from `movies` table.

---

## 7. Selecting Specific Columns

### Explanation

Instead of fetching all columns, it is better to fetch **only required columns** (performance benefit).

### Code

```sql
SELECT title, industry
FROM movies;
```

### Output

Shows only `title` and `industry` columns.

---

## 8. WHERE Clause (Filtering Rows)

### Explanation

`WHERE` is used to filter records based on a condition.

### Code

```sql
SELECT *
FROM movies
WHERE industry = "Bollywood";
```

### Output

Shows only Bollywood movies.

📌 SQL is **case-insensitive**.

---

## 9. COUNT Function

### Explanation

`COUNT(*)` returns the **number of rows**.

### Code

```sql
SELECT COUNT(*)
FROM movies
WHERE industry = "Hollywood";
```

### Output

Returns total Hollywood movies.

---

## 10. DISTINCT Keyword

### Explanation

`DISTINCT` returns **unique values**.

### Code

```sql
SELECT DISTINCT industry
FROM movies;
```

### Output

```
Bollywood
Hollywood
```

---

## 11. LIKE & Wildcard Search

### Explanation

`LIKE` is used for **pattern matching**.
`%` means **any number of characters**.

### Code

```sql
SELECT *
FROM movies
WHERE title LIKE "%Thor%";
```

### Output

All movies with **Thor** anywhere in title.

---

## 12. NULL Values

### Explanation

`NULL` means **missing or unknown value**.

### Code

```sql
SELECT *
FROM movies
WHERE studio IS NULL;
```

### Output

Movies with missing studio.

### Code (NOT NULL)

```sql
SELECT *
FROM movies
WHERE imdb_rating IS NOT NULL;
```

---

## 13. Comparison Operators

### Explanation

Used for numeric filtering.

| Operator | Meaning               |
| -------- | --------------------- |
| >        | Greater than          |
| <        | Less than             |
| >=       | Greater than or equal |
| <=       | Less than or equal    |

### Code

```sql
SELECT *
FROM movies
WHERE imdb_rating >= 9;
```

---

## 14. AND / OR Operators

### Explanation

* `AND`: all conditions must be true
* `OR`: any condition can be true

### Code (AND)

```sql
SELECT *
FROM movies
WHERE imdb_rating BETWEEN 6 AND 8;
```

### Code (OR)

```sql
SELECT *
FROM movies
WHERE release_year = 2018
   OR release_year = 2019
   OR release_year = 2020;
```

---

## 15. IN Operator

### Explanation

`IN` replaces multiple OR conditions.

### Code

```sql
SELECT *
FROM movies
WHERE release_year IN (2018, 2019, 2020);
```

---

## 16. ORDER BY Clause

### Explanation

Used to **sort results**.

### Code (Ascending – default)

```sql
SELECT *
FROM movies
ORDER BY imdb_rating;
```

### Code (Descending)

```sql
SELECT *
FROM movies
ORDER BY imdb_rating DESC;
```

---

## 17. LIMIT & OFFSET

### Explanation

* `LIMIT` → number of rows
* `OFFSET` → starting position (0-based index)

### Code

```sql
SELECT *
FROM movies
ORDER BY imdb_rating DESC
LIMIT 5;
```

### Code (Skip first row)

```sql
SELECT *
FROM movies
ORDER BY imdb_rating DESC
LIMIT 5 OFFSET 1;
```

---

## 18. Aggregate Functions

### Explanation

Used for **summary analytics**.

| Function | Purpose       |
| -------- | ------------- |
| COUNT    | Total rows    |
| MIN      | Minimum value |
| MAX      | Maximum value |
| AVG      | Average       |

### Code

```sql
SELECT 
MIN(imdb_rating) AS min_rating,
MAX(imdb_rating) AS max_rating,
AVG(imdb_rating) AS avg_rating
FROM movies
WHERE studio = "Marvel Studios";
```

---

## 19. ROUND Function

### Explanation

Used to limit decimal places.

### Code

```sql
SELECT ROUND(AVG(imdb_rating), 2) AS avg_rating
FROM movies;
```

---

## 20. GROUP BY Clause

### Explanation

Used to **group rows** and apply aggregate functions.

### Code

```sql
SELECT industry, COUNT(*) AS movie_count
FROM movies
GROUP BY industry;
```

### Output

Movies count per industry.

---

## 21. HAVING Clause

### Explanation

* `WHERE` → filters rows **before GROUP BY**
* `HAVING` → filters rows **after GROUP BY**

### Code

```sql
SELECT release_year, COUNT(*) AS movie_count
FROM movies
GROUP BY release_year
HAVING movie_count > 2;
```

---

## 22. Calculated / Derived Columns

### Explanation

Columns created using expressions.

### Code (Age Calculation)

```sql
SELECT *,
YEAR(CURDATE()) - birth_year AS age
FROM actors;
```

---

## 23. CASE Statement

### Explanation

Used for **multiple conditions**.

### Code

```sql
SELECT revenue,
unit,
CASE
  WHEN unit = "Billions" THEN revenue * 1000
  WHEN unit = "Thousands" THEN revenue / 1000
  ELSE revenue
END AS revenue_millions
FROM financials;
```

---

## 24. IF Condition

### Explanation

Used for **binary conditions**.

### Code

```sql
SELECT revenue,
IF(currency = "USD", revenue * 77, revenue) AS revenue_inr
FROM financials;
```

## 25. Why Multiple Tables Are Needed

### Explanation

Problems with single table:

1. **Data redundancy**
2. **Update difficulty**
3. **Poor organization**

Solution:

* Separate tables:

  * Customers
  * Items
  * Transactions
* Use **IDs** to connect tables

This reduces duplication and improves efficiency.
