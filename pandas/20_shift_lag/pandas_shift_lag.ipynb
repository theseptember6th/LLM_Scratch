```python
df["Prev_Price"]=df["Price"].shift(1)
df
```

```python
df["Price_Change"]=df["Price"]-df["Prev_Price"]
df
```

```python
df["5days"]=df["Price"].shift(5)
df
```

```python
df["5_day_%_return"]=((df["Price"]-df["5days"])/df["5days"])*100
df
```

```python
df=df[["Price"]]
df
```
