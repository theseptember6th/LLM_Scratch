1️⃣ Detect Google Colab

```python
import sys

if "google.colab" in sys.modules:
    print("Running on Google Colab kernel")
else:
    print("Running on Local / non-Colab kernel")
```

# Check runtime environment path

```python
import os

if "/content" in os.getcwd():
    print("Google Colab kernel")
else:
    print("Local kernel")
```

# Check GPU (strong hint)

```python
import subprocess

try:
    output = subprocess.check_output("nvidia-smi", shell=True).decode()
    print("GPU available → Likely Colab")
    print(output.split("\n")[0])
except:
    print("No GPU → Likely Local or Colab CPU")
```
