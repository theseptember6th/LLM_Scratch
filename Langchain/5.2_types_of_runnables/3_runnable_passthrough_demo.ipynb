RunnablePassthrough
- Purpose: A special primitive that passes its input through as output without any changes. It doesn't modify the data.

```python
from langchain_core.runnables import RunnablePassthrough

passthrough = RunnablePassthrough()
query=2
print(passthrough.invoke(query))  # o/p : 2

#what ever u invoke,same thing will be output
```

```python
print(passthrough.invoke({"name": "kristal"}))  # same dictionary as o/p
```
