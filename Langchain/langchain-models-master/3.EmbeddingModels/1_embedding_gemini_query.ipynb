```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from dotenv import load_dotenv
import os
load_dotenv()
api_key=os.getenv("GOOGLE_API_KEY")
embedding = GoogleGenerativeAIEmbeddings(
    model="gemini-embedding-001",  # or "models/text-embedding-004"
    task_type="retrieval_document",  # or "retrieval_query", "classification", etc.
    output_dimensionality=32
)

result = embedding.embed_query("Kathmandu is the capital of Nepal")

print(result)
```

```python
len(result)
```

**Task type**: Optional but recommended for better performance
**Dimensions**: Default by Gemini (768 dimensions for embedding-001)

**Available task types:**
- `"retrieval_document"` - for storing documents
- `"retrieval_query"` - for search queries
- `"classification"` - for classification tasks
- `"clustering"` - for clustering tasks

**Requirements:**
- `pip install langchain-google-genai`
- `GOOGLE_API_KEY` in your `.env` file

The output will be a list of floats representing the embedding vector.
