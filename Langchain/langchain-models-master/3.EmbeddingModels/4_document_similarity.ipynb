```python
from langchain_huggingface import HuggingFaceEmbeddings

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
# Set HuggingFace cache directory
import os

os.environ["HF_HOME"] = "/Users/kristalshrestha/Documents/Code/LLM_Scratch/models"

embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2") #fixed 384 dimensions
```

```python
# our external database lets say
documents = [
    "Virat Kohli is an Indian cricketer known for his aggressive batting and leadership.",
    "MS Dhoni is a former Indian captain famous for his calm demeanor and finishing skills.",
    "Sachin Tendulkar, also known as the 'God of Cricket', holds many batting records.",
    "Rohit Sharma is known for his elegant batting and record-breaking double centuries.",
    "Jasprit Bumrah is an Indian fast bowler known for his unorthodox action and yorkers.",
]
```

```python
# user query
query = "tell me about Virat Kohli"
```

```python
# Generate embeddings of both database and user query
doc_embeddings=embedding.embed_documents(documents)
query_embeddings=embedding.embed_query(query)
```

```python
print(cosine_similarity(X=[query_embeddings],Y=doc_embeddings))
```

### it return list of list,but always have one element(first element) so we take that element only

```python
print(cosine_similarity(X=[query_embeddings],Y=doc_embeddings)[0])
```

```python
scores=cosine_similarity(X=[query_embeddings],Y=doc_embeddings)[0]
```
