```python
from langchain_huggingface import HuggingFaceEmbeddings
import os

# Set HuggingFace cache directory
os.environ["HF_HOME"] = "/Users/kristalshrestha/Documents/Code/LLM_Scratch/models"

embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2") #fixed 384 dimensions
```

# for single text

```python
text = "Delhi is the capital of India"

vector = embedding.embed_query(text)
print(vector)
```

```python
# u can check embedding dimension like this also
len(vector)
```

# for multiple strings

```python
documents = [
    "kathmandu is the capital of Nepal",
    "Pokhara is the capital of province 4",
    "Kristal is my Name",
]

vector = embedding.embed_documents(documents)
print(vector)
```
