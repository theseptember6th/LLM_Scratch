```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv
import os

# Load environment variables
load_dotenv()

api_key = os.getenv("Hugging_face_api_token")

# Create LLM endpoint
llm = HuggingFaceEndpoint(
    # repo_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    repo_id="mistralai/Mistral-7B-Instruct-v0.2",
    task="text-generation",
    huggingfacehub_api_token=api_key,
    temperature=0, # Deterministic,Factual
    max_new_tokens=200
)

# Wrap with chat interface
model = ChatHuggingFace(llm=llm)

# Ask question
result = model.invoke("Write a 5 line poem on cricket game")

print(result.content)
```

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv
import os

# Load environment variables
load_dotenv()

api_key = os.getenv("Hugging_face_api_token")

# Create LLM endpoint
llm = HuggingFaceEndpoint(
    # repo_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    repo_id="mistralai/Mistral-7B-Instruct-v0.2",
    task="text-generation",
    huggingfacehub_api_token=api_key,
    temperature=1.0, # creative
    max_new_tokens=200 ## Model can generate up to 200 tokens
)

# Wrap with chat interface
model = ChatHuggingFace(llm=llm)

# Ask question
result = model.invoke("Write a 5 line poem on cricket game")

print(result.content)
```

# to check the values
