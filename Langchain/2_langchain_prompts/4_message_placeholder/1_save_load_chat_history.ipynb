#### saving this chat_history per session to a json file for now,in production you will save it in database

####### saving in a structured json,so later we can recreate Langchain message objects when loading

```python
import json

def save_chat(chat_history):
    data = []
    for msg in chat_history:
        data.append({
            "type": msg.__class__.__name__,
            "content": msg.content
        })

    with open("1chat_history.json", "w") as f:
        json.dump(data, f)
```

```python
save_chat(chat_history)
```

### loading back the chathistory into message objects

```python
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
import json

def load_chat():
    chat_history_path=r"./1chat_history.json"
    with open(chat_history_path, "r") as f:
        data = json.load(f)

    chat_history = []
    for msg in data:
        if msg["type"] == "HumanMessage":
            chat_history.append(HumanMessage(content=msg["content"]))
        elif msg["type"] == "AIMessage":
            chat_history.append(AIMessage(content=msg["content"]))
        elif msg["type"] == "SystemMessage":
            chat_history.append(SystemMessage(content=msg["content"]))

    return chat_history
```

```python
chat_history=load_chat()
```
