```python
import json
from langchain_core.messages import SystemMessage

def save_chat(chat_history):
    data = []
    for msg in chat_history:
        # Skip SystemMessage
        if isinstance(msg, SystemMessage):
            continue
        
        data.append({
            "type": msg.__class__.__name__,
            "content": msg.content
        })

    with open("1.1_no_systemMessage_chat_history.json", "w") as f:
        json.dump(data, f,indent=2)
```

```python
save_chat(chat_history)
```

```python
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage
import json

def load_chat():
    chat_history_path=r"./1.1_no_systemMessage_chat_history.json"
    with open(chat_history_path, "r") as f:
        data = json.load(f)

    chat_history = []
    for msg in data:
        if msg["type"] == "AIMessage":
            chat_history.append(AIMessage(content=msg["content"]))
        elif msg["type"] == "SystemMessage":
            chat_history.append(SystemMessage(content=msg["content"]))

    return chat_history
```

```python
chat_history_without_system_message=load_chat()
```
