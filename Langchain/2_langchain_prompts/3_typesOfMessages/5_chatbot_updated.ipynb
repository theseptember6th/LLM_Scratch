# using hugging face

```python
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint
from dotenv import load_dotenv
import os

# Load environment variables
load_dotenv()

api_key = os.getenv("Hugging_face_api_token")

# Create LLM endpoint
llm = HuggingFaceEndpoint(
    # repo_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    repo_id="mistralai/Mistral-7B-Instruct-v0.2",
    task="text-generation",
    huggingfacehub_api_token=api_key,
    max_new_tokens=10, # just to test
    temperature=0.5
)

# Wrap with chat interface
model = ChatHuggingFace(llm=llm)
```

```python
from langchain_core.messages import SystemMessage,AIMessage,HumanMessage

chat_history=[SystemMessage(content="You are a helpful assistant")] # to store chat history,but this time it labels who say who
while True:
    user_input=input("You :")
    chat_history.append(HumanMessage(user_input))
    if user_input.strip().lower() =="exit":
        break
    result=model.invoke(chat_history)
    chat_history.append(AIMessage(content=result.content))
    print(f"AI: {result.content}")
```

```python
print(chat_history)
```

# using Gemini


```python
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
from langchain_core.messages import AIMessage, SystemMessage, HumanMessage
import os
load_dotenv()
api_key=os.getenv("GOOGLE_API_KEY")

model = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite", google_api_key=api_key,
)
chat_history = [SystemMessage(content="You are helpful AI assistant")]
while True:
    user_input = input("You: ")
    chat_history.append(HumanMessage(content=user_input))
    if user_input.strip().lower() == "exit":
        break
    result = model.invoke(chat_history)
    chat_history.append(AIMessage(content=result.content))
    print(f"AI: {result.content}")


print(chat_history)
```
