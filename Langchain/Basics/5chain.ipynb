```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import BaseOutputParser
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import os
```

```python
api_key=os.getenv("OPENAI_API_KEY")
```

```python
chat_model=ChatOpenAI(
    model="gpt-4o-mini",api_key=api_key)
```

```python
class CommaSeparatedListOutputParser(BaseOutputParser):
    def parse(self,text:str):
        return text.strip().split(", ")

```

```python
template="""You are a helpful assistant who generates comma separated lists.A user will pass in a category,and you should generate 5 objects in that catetgory in a comma separated list.ONLY return a comma separated list, and nothing more."""
human_template="{text}"
```

```python
chat_prompt=ChatPromptTemplate.from_messages([
    ("system",template),
    ("human",human_template),
])
```

```python
chain=chat_prompt | chat_model | CommaSeparatedListOutputParser()

result=chain.invoke({"text":"colors"})
print(result)
```
