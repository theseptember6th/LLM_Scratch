### B. PyPDF Loader
- Used for loading PDF files, processing them page by page.
- Explanation: PyPDFLoader is perfect for standard, text-based PDFs. It reads the
PDF and creates one Document object for each page. This is very useful because
you can work with or analyze individual pages later.

##### Import

```python
from langchain_community.document_loaders import PyPDFLoader
```

##### Create loaded with the path to your file

```python
loader=PyPDFLoader(file_path="./2_dl_curriculum.pdf")
```

### load all pages

```python
docs=loader.load()
print(f"Total pages loaded: {len(docs)}") # same no of pages in your pdf
```

```python
print(docs)
```

```python
print(type(docs))
print(type(docs[0]))
```

##### acess the first page

```python
first_page=docs[0]
print(f"Metadata of page 1:{first_page.metadata}")
print(f"\n first 300 characters of content on that page: \n {first_page.page_content[:300]}")
```
