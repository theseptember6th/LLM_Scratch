##### Import

```python
from langchain_community.document_loaders import DirectoryLoader,PyPDFLoader
```

##### Create a loader for all PDFS inside a directory

```python
loader=DirectoryLoader(
    path="./3_books_directory_allpdfs", #Path to the directory
    glob="*.pdf", #Pattern to match PDF files
    loader_cls=PyPDFLoader #The loader to use for each file
)
```

##### Load all the documents

```python
docs=loader.load()
```

##### This total is the sum of all pages across all PDFs in the folder

```python
print(f"Total Document Objects(pages) loaded :{len(docs)}")
```
