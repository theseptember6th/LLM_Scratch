#### Import

```python
from langchain_text_splitters import CharacterTextSplitter

from langchain_community.document_loaders import PyPDFLoader
```

#### load the pdf

```python
loader=PyPDFLoader(
    file_path="./1.1_dl_curriculum.pdf"
)
docs=loader.load()
```

#### define the splitter object

```python
splitter=CharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=10,# for RAG based application,10% of chunksize for  chunk overlap is good.
    separator=""
)
```

```python
chunks=splitter.split_documents(docs)
```
